Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Plots that we can plot using the Seaborn library are:

Line plot---->This Plot Helps Us To See The Relationship Between X-Axis, Y-Axis And It Also Takes Some Parameters Such As Hue, Size, Color, Etc. 

Countplot---->>A Count Plot Is Used To Show The Counts Of Observations In Each Categorical Bin Using Bars.

Bar chart--->>A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. The Bar Chart Displays Data Using Several Bars, Each Representing A Particular Category. 

Pairplot--->>Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column.

Scatter plot---->>Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape. The Position Of Each Marker On The Horizontal And Vertical Axis Indicates Values For An Individual Data Point. This Plot Is Used To Observe Relationships Between Variables.

Histogram---->>A Histogram Is A Graph That Shows The Underlying Frequency Distribution Of A Set Of Continuous Data. 


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.

Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')
print(fmri.head())

sns.lineplot(x="timepoint", y="signal", data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri,hue="region")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri,hue="event",style="region")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. 

Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
print(titanic.head())


sns.boxplot(x="pclass", y="age", data=titanic)
plt.show()
```

```python
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column.

Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
diamonds = sns.load_dataset('diamonds')
print(diamonds.head())
```

```python
sns.histplot(data = diamonds,x = "price", hue="cut",bins=15,kde = False)
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot.

Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize=(15, 7))
sns.set_theme("poster")
data = np.random.rand(3, 3)
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
plt.show()
```
